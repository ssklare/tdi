```python
df = pd.read_csv('./Incidents_Responded_to_by_Fire_Companies.csv')
df['INCIDENT_DATE_TIME'] = pd.to_datetime(df['INCIDENT_DATE_TIME'])
```

```python
def time_bins(df):
    t_bins = np.zeros([24])
    c_bins = np.zeros([24])
    for i in np.arange(len(df)):
        bin_ind = df.at[i,'INCIDENT_DATE_TIME'].hour
        if df.at[i,'INCIDENT_TYPE_DESC'] == '113 - Cooking fire, confined to container':
            c_bins[bin_ind] = c_bins[bin_ind] + 1
        t_bins[bin_ind] = t_bins[bin_ind] + 1
    return t_bins, c_bins
```

```python
(total_bins, cook_bins) = time_bins(df)
```

```python
cook_fire_prop = (cook_bins/total_bins)
print(max(cook_fire_prop))
print(np.argmax(cook_fire_prop))
ans5 = np.argmax(cook_fire_prop)
```

```python
fire_array = fire_array.get_values()
fire_array = fire_array/sum(fire_array)
```

```python
def calc_r2(v_exp,v_obs):
    mean_obs = np.mean(v_obs)
    sstot = 0
    ssres = 0
    for i in np.arange(len(v_exp)):
        if v_exp[i]==0.0 or v_obs[i]==0.0:
            pass
        else:
            sstot = sstot + ((v_obs[i] - mean_obs)**2)
    #         print(sstot)
            ssres = ssres + ((v_obs[i] - v_exp[i])**2)
    #         print(ssres)
    r2 = 1 - (ssres/sstot)
    return r2
```

```python
calc_r2(v_exp=pop_array, v_obs=fire_array)
```

```python
yes_v = np.zeros([5])
no_v = np.zeros([5])
```

```python
for i in np.arange(len(df)):
    tm = (df.at[i,'TOTAL_INCIDENT_DURATION'])/60.0 # convert from seconds to minutes
#     print(tm)
    if df.at[i,'CO_DETECTOR_PRESENT_DESC'] == 'Yes':
        if tm >= 20 and tm < 30:
            yes_v[0] = yes_v[0] + 1
        elif tm >= 30 and tm < 40:
            yes_v[1] = yes_v[1] + 1
        elif tm >= 40 and tm < 50:
            yes_v[2] = yes_v[2] + 1
        elif tm >= 50 and tm < 60:
            yes_v[3] = yes_v[3] + 1
        elif tm >= 60 and tm <= 70:
            yes_v[4] = yes_v[4] + 1
    else:
        if tm >= 20 and tm < 30:
            no_v[0] = no_v[0] + 1
        elif tm >= 30 and tm < 40:
            no_v[1] = no_v[1] + 1
        elif tm >= 40 and tm < 50:
            no_v[2] = no_v[2] + 1
        elif tm >= 50 and tm < 60:
            no_v[3] = no_v[3] + 1
        elif tm >= 60 and tm <= 70:
            no_v[4] = no_v[4] + 1
#         print(i)
```
