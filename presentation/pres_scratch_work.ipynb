```python
# for creating data from from tag folders
def stls2df(stl_folder):
    stl_columns = ['name','facets','volume','cog' , 'inertia','tag']
    stl_df = pd.DataFrame( columns=stl_columns )
#     files = os.listdir('./tag1/')
    files = os.listdir('/Users/charlie/Dropbox/TDI/question3/' + stl_folder +'/')

    for i in np.arange(len(files)):  
        print(files[i])
        mesh1 = mesh.Mesh.from_file('/Users/charlie/Dropbox/TDI/question3/'+ stl_folder+'/' + files[i])
        stl_df.at[i,'name'] = files[i]
        stl_df.at[i,'facets'] = len(mesh1.normals)
        stl_df.at[i,'volume'], stl_df.at[i,'cog'],stl_df.at[i,'inertia'] = mesh1.get_mass_properties()
    return stl_df
```
